```python
def load_lifetimes_from_json(json_file):
    with open(json_file, 'r') as f:
        lifetimes = json.load(f)
    lifetimes = {int(resid): lifetime_list for resid, lifetime_list in lifetimes.items()}
    return lifetimes

def plot_kde_exponential(lifetime_data, save_fig=False, output_dir="", bw_adjust=1.5):
    plt.figure(figsize=(5, 2.6))
    y_positions = []
    min_vertical_gap = 0.1

    for lifetimes, label, color in lifetime_data:
        all_lifetimes = [lt * 10 for res_lifetimes in lifetimes.values() for lt in res_lifetimes]
        
        if not all_lifetimes:
            print(f"No lifetimes to plot for {label}.")
            continue
        average_lifetime = np.mean(all_lifetimes)
        sns.kdeplot(
            all_lifetimes, 
            bw_adjust=bw_adjust, 
            color=color, 
            label=label, 
            fill=True,
            log_scale=True,  
            alpha=0.1
        )

        plt.axvline(average_lifetime, color=color, linestyle='--', linewidth=2)

        y_pos = plt.gca().get_ylim()[1] * 0.9 
        while any(abs(y_pos - ypos) < min_vertical_gap for ypos in y_positions):
            y_pos -= min_vertical_gap 
        y_positions.append(y_pos) 

        plt.text(
            average_lifetime, 
            y_pos,  
            f'Avg: {average_lifetime:.2f}', 
            color=color, 
            ha='left',
            fontsize=12,
            bbox=dict(facecolor='lightgray', edgecolor='none', alpha=1, pad=0.2)
        )

    plt.xlabel('Lifetime (ps)')
    plt.ylabel('Density')
    plt.xlim(10, None)
    plt.title('SURF-TG lifetime')

    plt.legend(loc='upper left', bbox_to_anchor=(1, 1), borderaxespad=0)
    plt.tight_layout()

    if save_fig:
        if not os.path.exists(output_dir):
            os.makedirs(output_dir)
        output_path = os.path.join(output_dir, "trio_lifetimes_kde_buffer.png")
        plt.savefig(output_path, dpi=500, bbox_inches='tight')
        print(f"Saved KDE plot to {output_path}")
    
    plt.show()

if __name__ == "__main__":
    json_files = [
        'output_lifetime_c36/trio_lifetimes.json', 
        'output_lifetime_c36c/trio_lifetimes.json', 
        'output_lifetime_eq/trio_lifetimes.json'
    ]  

    labels = ['C36', 'C36c', 'Drude']
    colors = ['blue', 'red', 'green']
    
    lifetime_data = []
    for json_file, label, color in zip(json_files, labels, colors):
        if not os.path.isfile(json_file):
            print(f"File not found: {json_file}")
            continue
        lifetimes = load_lifetimes_from_json(json_file)
        lifetime_data.append((lifetimes, label, color))

    plot_kde_exponential(
        lifetime_data, 
        save_fig=True,            
        output_dir='output_plots',  
        bw_adjust=1               
    )
```

```python

def plot_ecdf(lifetimes, save_fig=False, output_dir=""):
    all_lifetimes = np.sort([lt for res_lifetimes in lifetimes.values() for lt in res_lifetimes])
    
    if not all_lifetimes.size:
        print("No lifetimes to plot.")
        return
    
    y = np.arange(1, len(all_lifetimes)+1) / len(all_lifetimes)
    
    plt.figure(figsize=(5, 3))
    plt.step(all_lifetimes, y, where='post', color='skyblue', label='ECDF')
    
    average_lifetime = np.mean(all_lifetimes)
    plt.axvline(average_lifetime, color='red', linestyle='dashed', linewidth=2, label=f'Average: {average_lifetime:.2f}')
    plt.text(average_lifetime + 0.05, 0.5, f'Average: {average_lifetime:.2f}', color='red')
    
    plt.xlabel('Lifetime (frames)')
    plt.ylabel('ECDF')
    plt.xscale('log')
    plt.title('Empirical Cumulative Distribution Function of TRIO Molecule Lifetimes')
    plt.xlim(-10,3000)
    plt.legend()
    
    if save_fig:
        if not os.path.exists(output_dir):
            os.makedirs(output_dir)
        output_path = os.path.join(output_dir, "trio_lifetimes_ecdf.png")
        plt.savefig(output_path, dpi=300, bbox_inches='tight')
        print(f"Saved ECDF plot to {output_path}")
    
    plt.show()

# Example usage:
if __name__ == "__main__":
    json_file = 'output_lifetime_c36/trio_lifetimes.json'
    lifetimes = load_lifetimes_from_json(json_file)
    plot_ecdf(lifetimes, save_fig=False, output_dir='output_lifetime_c36')
```

```python
if __name__ == "__main__":
    json_file = 'output_lifetime_c36c/trio_lifetimes.json'
    lifetimes = load_lifetimes_from_json(json_file)
    plot_ecdf(lifetimes, save_fig=False, output_dir='output_lifetime_c36c')
```
