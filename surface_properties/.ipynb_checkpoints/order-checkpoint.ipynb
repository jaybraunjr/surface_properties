```python
def load_and_plot(ax, folders, filenames, labels, title, linewidth=2):

    for folder, label in zip(folders, labels):
        data = []
        for filename in filenames:
            data.append(np.loadtxt(f'{folder}/{filename}_bottom.dat'))
            data.append(np.loadtxt(f'{folder}/{filename}_top.dat'))
        avg_data = np.mean(data, axis=0)
        ax.plot(avg_data[:, 0], avg_data[:, 1], label=label, linewidth=linewidth)
    ax.set_title(title)
    ax.set_xlabel('Carbon number')
    ax.set_ylabel('Order parameter')
    ax.legend(loc='best')

fig, ax = plt.subplots(figsize=(6, 3))

drude_folders = ['drude']
drude_labels = ['drude test']
filenames = ['DOPE2']
load_and_plot(ax, drude_folders, filenames, drude_labels, 'DOPE Order Parameters (Drude)', linewidth=2)

plt.tight_layout(rect=[0, 0, 0.85, 1])  # Adjust layout to fit the legend
# plt.savefig('DOPE.png', dpi=500, bbox_inches='tight')  # Save with better spacing
plt.show()
```

```python
from collections import defaultdict
import numpy as np
import MDAnalysis as mda
import os
import json
from tqdm import tqdm

class OrderParameters:
    def __init__(self, u, atomlists, resname, selection, start_frame=0):        
        self.u = u
        self.atomlists = atomlists
        self.resname = resname
        self.selection = selection
        self.start_frame = start_frame
        self.C_numbers, self.Cs, self.Hs_f, self.repeat = self.process_atom_lists()

    def process_atom_lists(self):
        C_numbers = []
        Cs = []
        Hs = []
        repeat = []
        
        for atoms in self.atomlists:
            C_number = atoms[0][2:]
            C_numbers.append(int(C_number))
            Cs.append(atoms[0])
            Hs.append(atoms[1:])
            repeat.append(len(atoms) - 1)
        
        Hs_f = [item for sublist in Hs for item in sublist]
        
        assert int(np.sum(repeat)) == len(Hs_f), "Mismatch in repeats"
        return C_numbers, Cs, Hs_f, repeat



    def compute_OP(self):
        print("Selection string:", self.selection)
        all_molecules = self.u.select_atoms(self.selection, updating=True)
        if len(all_molecules) == 0:
#             print("No atoms selected for this configuration; skipping frame.")
            return np.transpose([self.C_numbers, np.full(len(self.C_numbers), np.nan)])

        output = []
        total_frames = len(self.u.trajectory)

        with tqdm(total=total_frames - self.start_frame, desc="Processing frames") as pbar:
            for idx, ts in enumerate(self.u.trajectory[self.start_frame:]):
                valid_indices_group1 = []
                valid_indices_group2 = []

                for molecule in all_molecules.residues:
                    atoms_in_molecule = molecule.atoms
                    if all(atom.index in all_molecules.indices for atom in atoms_in_molecule):
                        valid_indices_group1.extend(molecule.atoms.select_atoms("name " + " ".join(self.Cs)).indices)
                        valid_indices_group2.extend(molecule.atoms.select_atoms("name " + " ".join(self.Hs_f)).indices)

#                 # Check if selections are non-empty before proceeding
#                 if not valid_indices_group1 or not valid_indices_group2:
#                     print("No valid indices found for selected atoms; skipping frame.")
#                     continue

                group1 = self.u.atoms[valid_indices_group1]
                group2 = self.u.atoms[valid_indices_group2]

                natoms = len(self.Cs)
                nmols = int(len(group1.positions) / natoms)
                repeats = self.repeat * nmols

                p1 = np.repeat(group1.positions, repeats, axis=0)
                p2 = group2.positions
                dp = p2 - p1
                norm = np.sqrt(np.sum(np.power(dp, 2), axis=-1))

                # Avoid division by zero
                cos_theta = np.divide(dp[..., 2], norm, out=np.zeros_like(dp[..., 2]), where=norm!=0)

                S = -0.5 * (3 * np.square(cos_theta) - 1)
                new_S = self._average_over_hydrogens(S, repeats)
                new_S.shape = (nmols, natoms)
                results = np.average(new_S, axis=0)
                output.append(results)
                
                pbar.update(1)

        avg = np.average(output, axis=0)
        return np.transpose([self.C_numbers, avg])



    def _average_over_hydrogens(self, x, reps):
        assert len(x) == int(np.sum(reps)), 'Mismatch in repeats'
        i = 0
        out = []
        for rep in reps:
            tmp = []
            for r in range(rep):
                tmp.append(x[i])
                i += 1
            out.append(np.average(tmp))
        return np.array(out)

# def run_op(TOP, TRAJ, output_folder_name, lipid_info, start_frame, name, use_ranges=False, ranges=None):
#     u = mda.Universe(TOP, TRAJ)
#     output_folder = output_folder_name
#     if not os.path.exists(output_folder):
#         os.makedirs(output_folder)
#     leaflets = ["bottom", "top"]
#     halfz = u.dimensions[2] / 2

#     # Check if protein is present on each leaflet
#     has_protein_top = len(u.select_atoms(f"protein and prop z > {halfz}")) > 0
#     has_protein_bottom = len(u.select_atoms(f"protein and prop z < {halfz}")) > 0

#     for leaflet in leaflets:
#         print(f"\nProcessing {leaflet} leaflet ...")
#         if leaflet == "bottom":
#             z_condition = f"prop z < {halfz}"
#             has_protein = has_protein_bottom
#             for lipid, atomlist in lipid_info.items():
#                 selection = f"resname {lipid} and {z_condition}"
#                 print(f"Using selection for bottom leaflet: {selection}")
#                 OP = OrderParameters(u, atomlist, lipid, selection, start_frame)
#                 results = OP.compute_OP()
#                 output_path = os.path.join(output_folder, f"{name}_{leaflet}_control.dat")
#                 np.savetxt(output_path, results)
#                 print(f"Saved to {output_path}\n")

#         else:
#             z_condition = f"prop z > {halfz}"
#             has_protein = has_protein_top
#             if use_ranges and ranges and has_protein:
#                 for start, end, label in ranges:
#                     for lipid, atomlist in lipid_info.items():
#                         selection = f"resname {lipid} and (around {end} protein) and not (around {start} protein) and {z_condition}"
#                         print(f"Using selection for top leaflet with range {start}-{end} nm: {selection}")
#                         OP = OrderParameters(u, atomlist, lipid, selection, start_frame)
#                         results = OP.compute_OP()
#                         output_path = os.path.join(output_folder, f"{name}_{leaflet}_{label}nm.dat")
#                         np.savetxt(output_path, results)
#                         print(f"Saved to {output_path}\n")
#             else:
#                 for lipid, atomlist in lipid_info.items():
#                     selection = f"resname {lipid} and {z_condition}"
#                     print(f"Using selection for top leaflet without ranges: {selection}")
#                     OP = OrderParameters(u, atomlist, lipid, selection, start_frame)
#                     results = OP.compute_OP()
#                     output_path = os.path.join(output_folder, f"{name}_{leaflet}_control.dat")
#                     np.savetxt(output_path, results)
#                     print(f"Saved to {output_path}\n")

```

```python
def run_op(TOP, TRAJ, output_folder_name, lipid_info, start_frame, name, use_ranges=False, ranges=None):
    u = mda.Universe(TOP, TRAJ)
    output_folder = output_folder_name
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    leaflets = ["bottom", "top"]
    halfz = u.dimensions[2] / 2

    # Check if protein is present on each leaflet
    has_protein_top = len(u.select_atoms(f"protein and prop z > {halfz}")) > 0
    has_protein_bottom = len(u.select_atoms(f"protein and prop z < {halfz}")) > 0

    for leaflet in leaflets:
        print(f"\nProcessing {leaflet} leaflet ...")
        if leaflet == "bottom":
            z_condition = f"prop z < {halfz}"
            has_protein = has_protein_bottom
            for lipid, atomlist in lipid_info.items():
                selection = f"resname {lipid} and {z_condition}"
                print(f"Using selection for bottom leaflet: {selection}")
                OP = OrderParameters(u, atomlist, lipid, selection, start_frame)
                results = OP.compute_OP()
                output_path = os.path.join(output_folder, f"{name}_{leaflet}_control.dat")
                np.savetxt(output_path, results)
                print(f"Saved to {output_path}\n")

#         else:
#             z_condition = f"prop z > {halfz}"
#             has_protein = has_protein_top
#             if use_ranges and ranges and has_protein:
#                 for start, end, label in ranges:
#                     for lipid, atomlist in lipid_info.items():
#                         # Temporarily simplify selection for testing
#                         selection = f"resname {lipid} and (around {end} protein) and {z_condition}"
#                         print(f"Using simplified selection for top leaflet with range up to {end} nm: {selection}")
#                         OP = OrderParameters(u, atomlist, lipid, selection, start_frame)
#                         results = OP.compute_OP()
#                         output_path = os.path.join(output_folder, f"{name}_{leaflet}_{label}nm.dat")
#                         np.savetxt(output_path, results)
#                         print(f"Saved to {output_path}\n")
#             else:
#                 for lipid, atomlist in lipid_info.items():
#                     selection = f"resname {lipid} and {z_condition}"
#                     print(f"Using selection for top leaflet without ranges: {selection}")
#                     OP = OrderParameters(u, atomlist, lipid, selection, start_frame)
#                     results = OP.compute_OP()
#                     output_path = os.path.join(output_folder, f"{name}_{leaflet}_control.dat")
#                     np.savetxt(output_path, results)
#                     print(f"Saved to {output_path}\n")

            else:
                z_condition = f"prop z > {halfz}"
                has_protein = has_protein_top
                if use_ranges and ranges and has_protein:
                    for start, end, label in ranges:
                        for lipid, atomlist in lipid_info.items():
                            selection = f"resname {lipid} and (around {end} protein) and not (around {start} protein) and {z_condition}"
                            print(f"Using selection for top leaflet with range {start}-{end} nm: {selection}")
                            OP = OrderParameters(u, atomlist, lipid, selection, start_frame)
                            results = OP.compute_OP()
                            output_path = os.path.join(output_folder, f"{name}_{leaflet}_{label}nm.dat")
                            np.savetxt(output_path, results)
                            print(f"Saved to {output_path}\n")
                else:
                    for lipid, atomlist in lipid_info.items():
                        selection = f"resname {lipid} and {z_condition}"
                        print(f"Using selection for top leaflet without ranges: {selection}")
                        OP = OrderParameters(u, atomlist, lipid, selection, start_frame)
                        results = OP.compute_OP()
                        output_path = os.path.join(output_folder, f"{name}_{leaflet}_control.dat")
                        np.savetxt(output_path, results)
                        print(f"Saved to {output_path}\n")

```

```python
def load_and_plot(ax, folders, filenames, labels, title, linewidth=2):

    for folder, label in zip(folders, labels):
        data_bottom = []
        data_top = []
        
        # Load both top and bottom data for each filename
        for filename in filenames:
            data_bottom.append(np.loadtxt(f'{folder}/{filename}_bottom_control.dat'))
            data_top.append(np.loadtxt(f'{folder}/{filename}_top_5to10nm.dat'))
        
        # Average data across all files for bottom and top separately
        avg_data_bottom = np.mean(data_bottom, axis=0)
        avg_data_top = np.mean(data_top, axis=0)

        # Plot each leaflet separately
        ax.plot(avg_data_bottom[:, 0], avg_data_bottom[:, 1], label=f"{label} Bottom", linewidth=linewidth, linestyle='-')
        ax.plot(avg_data_top[:, 0], avg_data_top[:, 1], label=f"{label} Top", linewidth=linewidth, linestyle='--')

    # Set plot title and labels
    ax.set_title(title)
    ax.set_xlabel('Carbon number')
    ax.set_ylabel('Order parameter')
    ax.legend(loc='best')

# Example usage
fig, ax = plt.subplots(figsize=(6, 3))

folders = ['mlx_protein']
labels = ['mlx']
filenames = ['DOPE2']
load_and_plot(ax, folders, filenames, labels, 'DOPE Order Parameters MLX', linewidth=2)

plt.tight_layout(rect=[0, 0, 0.85, 1])  # Adjust layout to fit the legend
# plt.savefig('DOPE.png', dpi=500, bbox_inches='tight')  # Save with better spacing
plt.show()
```
